Importation des Données

```python
import pandas as pd
import numpy as np

df = pd.read_excel("../data/credit.xls", header=1)
print(df.shape)
df.head()
```

Renommer les colonnes

```python
df = df.rename(columns={
    'default payment next month': 'default',
    'PAY_0': 'PAY_1'  # correction typo UCI
})
df.columns.tolist()
```

Supprimer la colonne ID

```python
df = df.drop(columns=['ID'])
```

Vérifier les valeurs aberrantes

```python
# EDUCATION contient des valeurs inconnues (0, 5, 6) non documentées
print(df['EDUCATION'].value_counts())
print(df['MARRIAGE'].value_counts())
```

Nettoyer EDUCATION et MARRIAGE

```python
# On regroupe les valeurs inconnues dans la catégorie "autre"
df['EDUCATION'] = df['EDUCATION'].replace({0: 4, 5: 4, 6: 4})
df['MARRIAGE'] = df['MARRIAGE'].replace({0: 3})

print(df['EDUCATION'].value_counts())
print(df['MARRIAGE'].value_counts())
```

Vérifier les valeurs manquantes

```python
df.isnull().sum()
```

Doublons

```python
print(f"Doublons : {df.duplicated().sum()}")
df = df.drop_duplicates()
```

Statistiques descriptives

```python
df.describe().T
```

Sauvegarde du dataset propre

```python
df.to_csv("../data/credit_clean.csv", index=False)
print("✅ Dataset propre sauvegardé !")
```
